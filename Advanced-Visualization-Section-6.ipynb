```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline        
#import numpy.ndarray
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("white",{"axes.facecolor": "black"}) # white, whitegrid, dark, darkgrid, ticks # we can change by passing keyward parameter in dictionary
f, axes = plt.subplots(2,2, figsize=(10,10))
# plot [0,0]
k1 = sns.kdeplot(movies.TotalBudget, movies.TotalCollection, \
                 shade= True, shade_lowest=True, cmap= 'inferno', \
                 ax= axes[0,0])
k1b = sns.kdeplot(movies.TotalBudget, movies.TotalCollection, \
                 cmap= 'cool', ax= axes[0,0])
# plot [0,1]
k2 = sns.kdeplot(movies.TotalCollection, movies.CriticRating, \
                 shade= True, shade_lowest=True, cmap= 'inferno', \
                 ax= axes[0,1])
k2b = sns.kdeplot(movies.TotalCollection, movies.CriticRating, \
                 cmap= 'cool', ax= axes[0,1])

# plot [1,0]
z = sns.violinplot(data=movies, x='Verdict', y='CriticRating',ax= axes[1,0], \
                   palette='YlOrRd')

# plot [1,1]
k4 = sns.kdeplot(movies.TotalBudget, movies.TotalCollection, shade=True , shade_lowest= False, cmap='Blues_r',\
                
                 ax=axes[1,1])

# plot [1,1]
k4b = sns.kdeplot(movies.TotalBudget, movies.TotalCollection, cmap='gist_gray_r', ax=axes[1,1])
#axes[1,1].hist(movies.TotalBudget, bins=15)
k1.set(xlim=(-20,160))
k2.set(xlim=(-60,300))
plt.show()
```

```python
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
lst= list()
mylabels= list()
for v in movies.Verdict.cat.categories:
    lst.append(movies[movies.Verdict == v].TotalCollection)
    mylabels.append(v)
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)  # size of A4 paper
plt.title("Movies Budget Distribution", Fontsize=35, color="Blue", fontname='Times new roman', loc="center")
plt.ylabel("Number of Movies", fontsize=25, color='green' )
plt.xlabel("Budget", fontsize=25, color='red' )
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
h=plt.hist(lst, label= mylabels, stacked=True)
#h=plt.hist(lst, label= mylabels, stacked=True, color=['red','blue','yellow','green','violet','purple','black','yellow','green','violet'])
plt.legend(frameon= True, shadow =True, fancybox= True, framealpha=1, prop={'size':20})
plt.show()
```
